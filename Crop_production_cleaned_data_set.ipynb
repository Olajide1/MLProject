```python
import pandas as pd

# Load the dataset
file_path = 'Crop_production.csv'
crop_data = pd.read_csv(file_path)

# Handle missing values
# Fill missing values with appropriate values (e.g., mean for numeric columns)
crop_data = crop_data.fillna({
    'P': crop_data['P'].mean(),
    'K': crop_data['K'].mean(),
    'pH': crop_data['pH'].mean(),
    'rainfall': crop_data['rainfall'].mean(),
    'temperature': crop_data['temperature'].mean(),
    'Area_in_hectares': crop_data['Area_in_hectares'].mean(),
    'Production_in_tons': crop_data['Production_in_tons'].mean(),
    'Yield_ton_per_hec': crop_data['Yield_ton_per_hec'].mean()
})

# Ensure correct data types
crop_data['State_Name'] = crop_data['State_Name'].astype(str)
crop_data['Crop_Type'] = crop_data['Crop_Type'].astype(str)
crop_data['Crop'] = crop_data['Crop'].astype(str)
# Ensure numeric columns are of numeric types
numeric_columns = ['P', 'K', 'pH', 'rainfall', 'temperature', 'Area_in_hectares', 'Production_in_tons', 'Yield_ton_per_hec']
crop_data[numeric_columns] = crop_data[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Normalize text (e.g., state names)
crop_data['State_Name'] = crop_data['State_Name'].str.strip().str.lower()

# Check for and handle duplicates
crop_data = crop_data.drop_duplicates()

# Save the cleaned dataset
cleaned_file_path = 'Cleaned_Crop_production.csv'
crop_data.to_csv(cleaned_file_path, index=False)

# Display the first few rows of the cleaned dataframe to verify
crop_data.head()
```
