```python
import pandas as pd

# Load the dataset
rainfall_data_file_path = 'Data_after_rainfall.csv'
rainfall_data = pd.read_csv(rainfall_data_file_path)

# Handle missing values
# Fill missing values with appropriate values (e.g., mean for numeric columns)
numeric_columns = ['rainfall', 'Area_in_hectares', 'Yield_ton_per_hec', 'Production_in_tons']
rainfall_data[numeric_columns] = rainfall_data[numeric_columns].apply(pd.to_numeric, errors='coerce')

rainfall_data = rainfall_data.fillna({
    'rainfall': rainfall_data['rainfall'].mean(),
    'Area_in_hectares': rainfall_data['Area_in_hectares'].mean(),
    'Yield_ton_per_hec': rainfall_data['Yield_ton_per_hec'].mean(),
    'Production_in_tons': rainfall_data['Production_in_tons'].mean()
})

# Ensure correct data types
rainfall_data['State_Name'] = rainfall_data['State_Name'].astype(str)
rainfall_data['Crop_Type'] = rainfall_data['Crop_Type'].astype(str)
rainfall_data['Crop'] = rainfall_data['Crop'].astype(str)
# Ensure numeric columns are of numeric types
rainfall_data[numeric_columns] = rainfall_data[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Normalize text (e.g., state names)
rainfall_data['State_Name'] = rainfall_data['State_Name'].str.strip().str.lower()
rainfall_data['Crop_Type'] = rainfall_data['Crop_Type'].str.strip().str.lower()
rainfall_data['Crop'] = rainfall_data['Crop'].str.strip().str.lower()

# Check for and handle duplicates
rainfall_data = rainfall_data.drop_duplicates()

# Save the cleaned dataset
cleaned_rainfall_file_path = 'Cleaned_Data_after_rainfall.csv'
rainfall_data.to_csv(cleaned_rainfall_file_path, index=False)

# Display the first few rows of the cleaned dataframe to verify
rainfall_data.head()
```
