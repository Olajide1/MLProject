```python
import pandas as pd

# Load the dataset
crop_yield_file_path = 'crop_yield.csv'
crop_yield_data = pd.read_csv(crop_yield_file_path)

# Handle missing values
# Fill missing values with appropriate values (e.g., mean for numeric columns)
crop_yield_data['Area'] = pd.to_numeric(crop_yield_data['Area'], errors='coerce')
crop_yield_data['Production'] = pd.to_numeric(crop_yield_data['Production'], errors='coerce')

crop_yield_data = crop_yield_data.fillna({
    'Area': crop_yield_data['Area'].mean(),
    'Production': crop_yield_data['Production'].mean()
})

# Ensure correct data types
crop_yield_data['State_Name'] = crop_yield_data['State_Name'].astype(str)
crop_yield_data['District_Name'] = crop_yield_data['District_Name'].astype(str)
crop_yield_data['Crop_Year'] = crop_yield_data['Crop_Year'].astype(int)
crop_yield_data['Season'] = crop_yield_data['Season'].astype(str)
crop_yield_data['Crop'] = crop_yield_data['Crop'].astype(str)
# Ensure numeric columns are of numeric types
numeric_columns = ['Area', 'Production']
crop_yield_data[numeric_columns] = crop_yield_data[numeric_columns].apply(pd.to_numeric, errors='coerce')

# Normalize text (e.g., state names)
crop_yield_data['State_Name'] = crop_yield_data['State_Name'].str.strip().str.lower()
crop_yield_data['District_Name'] = crop_yield_data['District_Name'].str.strip().str.lower()
crop_yield_data['Season'] = crop_yield_data['Season'].str.strip().str.lower()
crop_yield_data['Crop'] = crop_yield_data['Crop'].str.strip().str.lower()

# Check for and handle duplicates
crop_yield_data = crop_yield_data.drop_duplicates()

# Save the cleaned dataset
cleaned_yield_file_path = 'Cleaned_Crop_yield.csv'
crop_yield_data.to_csv(cleaned_yield_file_path, index=False)

# Display the first few rows of the cleaned dataframe to verify
crop_yield_data.head()
```
